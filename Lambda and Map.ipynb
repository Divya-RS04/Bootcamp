

### Map and Lambda functions


```python
def addition(x,y):
    return x + y

addition(2,3)
```

```python
numbers=lambda x:x**2
print(type(numbers))
print(numbers(12))
```

```python
numbers1=[1,2,3,4,5,6]
map(lambda x:x**2,numbers1)
```

```python
list(map(lambda x:x**2,numbers1))
```

```python
def get_name(person):
    return person['name']
people=[

    {'name':'Krish','age':32},
    {'name':'Naik','age':33}
]

list(map(get_name,people))
```

```python
words=["apple","banana","Cherry"]
words_upper=list(map(str.upper,words))
print(words_upper)
```

```python
str_numbers=['1','2','3','4','5']
int_numbers=list(map(int,str_numbers))
print(int_numbers)
```
